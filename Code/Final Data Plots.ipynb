```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tfidf_svm = [0.91,0.66, 0.76]
simple_cnn = [0.93,0.69,0.79]
c_cnn = [0.93,0.66, 0.78]
simple_lstm = [0.86,0.65,0.82]
cnn_gru = [0.93, 0.68, 0.83]
novel_cnn_lstm = [0.93,0.68,0.81]
```

```python


# Bar plot
plt.figure(figsize=(8,8))
labels = ['Davidson', 'Hateval', 'Ethos']
x = range(len(labels))
bar_width = 0.1  # Width of each bar
plt.bar(x, tfidf_svm, width=bar_width, label='TFIDF-SVM', color='blue')
plt.bar([i + bar_width for i in x], simple_cnn, width=bar_width, label='Simple_CNN', color='orange')
plt.bar([i + 2 * bar_width for i in x], c_cnn, width=bar_width, label='2-3-4 CNN', color='green')
plt.bar([i + 3 * bar_width for i in x], simple_lstm, width=bar_width, label='Simple_LSTM', color='firebrick')
plt.bar([i + 4 * bar_width for i in x], cnn_gru, width=bar_width, label='CNN_GRU', color='purple')
plt.bar([i + 5 * bar_width for i in x], novel_cnn_lstm, width=bar_width, label='2-3-4 CNN_LSTM', color='black')
plt.xlabel('Dataset')
plt.ylabel('F1 Score')
plt.title('Comparison of Deep Learning Models')
plt.xticks([i + bar_width for i in x], labels)
plt.legend()
plt.ylim(0, 1.0)
plt.show()
```

```python
standalone = [0.88,0.74, 0.82, 0.94, 0.92]
ensemble = [0.85,0.55, 0.81, 0.9, 0.88]
aug = [0.87,0.61,0.81,0.92,0.9]
d_ensemble = [0.85,0.67,0.78,0.9,0.89]
```

```python
# Bar plot
plt.figure(figsize=(12,6))
labels = ['Davidson', 'Hateval', 'Ethos', "Jigsaw", "Qian"]
x = range(len(labels))
bar_width = 0.2  # Width of each bar
plt.bar(x, standalone, width=bar_width, label='Standalone', color='blue')
plt.bar([i + bar_width for i in x], ensemble, width=bar_width, label='Stacked Ensemble', color='orange')
plt.bar([i + 2 * bar_width for i in x], aug, width=bar_width, label='Augmented Dataset', color='firebrick')
plt.bar([i + 3 * bar_width for i in x], d_ensemble, width=bar_width, label='Dynamic Ensemble', color='green')
plt.xlabel('Dataset')
plt.ylabel('F1 Score')
plt.title('Comparison of Proposed Models')
plt.xticks([i + bar_width for i in x], labels)
plt.legend(loc=(0.21,0.8))
plt.ylim(0, 1.0)
plt.show()
```
