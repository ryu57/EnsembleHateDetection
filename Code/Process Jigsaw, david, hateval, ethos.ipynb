```python
import numpy as np
import pandas as pd
import functools
import ast
import re
```

```python
df1 = pd.read_csv("literature/datasets/jigsaw/train.csv")
```

```python


def process1(data):
    data = re.sub(r'\n'," ",data)
    return data.strip()
    

def process_data(df):
    processed_column = df['comment_text']
    processed_column = processed_column.apply(lambda x:process1(x))
    df["text"] = processed_column
    return df
```

```python
df1 = process_data(df1)
```

```python
df_new = pd.DataFrame()
df_new["text"] = df1["text"]
df_new["class"] = df1["identity_hate"]
```

```python
df_new.to_csv("literature/datasets/jigsaw/processed.csv",header=True, index=False)
```

```python
datasets = {
    "davidson" : pd.read_csv("literature\datasets\dt\labeled_data_all_2classes_only.csv", index_col=0),
    "hateval" : pd.read_csv("literature\\datasets\\hateval\\train_en.tsv", sep='\t', index_col=0),
    "ethos" : pd.read_csv("literature\\datasets\\ethos\\Ethos_Dataset_Binary.csv", sep=';')
}
```

```python
datasets["davidson"] = datasets["davidson"].rename({"tweet":"text"}, axis=1)
datasets["davidson"]['class'] = datasets["davidson"]['class'].replace({0: 1, 2: 0})

datasets["hateval"] = datasets["hateval"].rename({"HS":"class"}, axis=1)

datasets["ethos"] = datasets["ethos"].rename({"comment":"text", "isHate":"class"}, axis=1)
datasets["ethos"]['class'] = datasets["ethos"]['class'].astype("int")
```

```python
df_david = pd.DataFrame()
df_david["text"] = datasets["davidson"]["text"]
df_david["class"] = datasets["davidson"]["class"]
df_david
```

```python
df_hateval = pd.DataFrame()
df_hateval["text"] = datasets["hateval"]["text"]
df_hateval["class"] = datasets["hateval"]["class"]
df_hateval
```

```python
df_ethos = pd.DataFrame()
df_ethos["text"] = datasets["ethos"]["text"]
df_ethos["class"] = datasets["ethos"]["class"]
df_ethos
```

```python
df_david.to_csv("datasets/model_training/davidson.csv",header=True, index=False)
df_hateval.to_csv("datasets/model_training/hateval.csv",header=True, index=False)
df_ethos.to_csv("datasets/model_training/ethos.csv",header=True, index=False)
```
